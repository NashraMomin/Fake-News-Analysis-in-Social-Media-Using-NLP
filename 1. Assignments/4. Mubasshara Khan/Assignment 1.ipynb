```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import cifar10
```

```python
# Load CIFAR-10 dataset
(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()

# Normalize the images to the range [0, 1]
train_images, test_images = train_images / 255.0, test_images / 255.0
```

```python
# Build the ANN model
ann_model = models.Sequential([
    layers.Flatten(input_shape=(32, 32, 3)),  # Flatten the 32x32x3 images
    layers.Dense(512, activation='relu'),     # Fully connected layer with 512 units
    layers.Dense(10, activation='softmax')    # Output layer with 10 classes
])

# Compile the model
ann_model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

# Train the model
history_ann = ann_model.fit(train_images, train_labels, epochs=10, validation_data=(test_images, test_labels))
```

```python
# Build the CNN model
cnn_model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),  # Conv layer with 32 filters
    layers.MaxPooling2D((2, 2)),  # Max pooling
    layers.Conv2D(64, (3, 3), activation='relu'),  # Conv layer with 64 filters
    layers.MaxPooling2D((2, 2)),  # Max pooling
    layers.Conv2D(64, (3, 3), activation='relu'),  # Conv layer with 64 filters
    layers.Flatten(),  # Flatten the 3D output to 1D
    layers.Dense(64, activation='relu'),  # Fully connected layer
    layers.Dense(10, activation='softmax')  # Output layer
])

# Compile the model
cnn_model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

# Train the model
history_cnn = cnn_model.fit(train_images, train_labels, epochs=10, validation_data=(test_images, test_labels))
```

```python
# Evaluate ANN model
ann_test_loss, ann_test_acc = ann_model.evaluate(test_images, test_labels)

# Evaluate CNN model
cnn_test_loss, cnn_test_acc = cnn_model.evaluate(test_images, test_labels)

# Print results
print(f'ANN Test Accuracy: {ann_test_acc:.4f}')
print(f'CNN Test Accuracy: {cnn_test_acc:.4f}')
```

```python
# Plot Accuracy
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(history_ann.history['accuracy'], label='ANN Training Accuracy')
plt.plot(history_ann.history['val_accuracy'], label='ANN Validation Accuracy')
plt.plot(history_cnn.history['accuracy'], label='CNN Training Accuracy')
plt.plot(history_cnn.history['val_accuracy'], label='CNN Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot Loss
plt.subplot(1, 2, 2)
plt.plot(history_ann.history['loss'], label='ANN Training Loss')
plt.plot(history_ann.history['val_loss'], label='ANN Validation Loss')
plt.plot(history_cnn.history['loss'], label='CNN Training Loss')
plt.plot(history_cnn.history['val_loss'], label='CNN Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
